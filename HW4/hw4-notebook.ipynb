```python
def remove_mean(data): #move to (0,0) aka mean center
    #nptable = data.to_numpy()
    mean_vec = np.mean(data, axis=0)
    table_std = data.copy() #shallow copy so technically i can return data or table_std (it doesnt matter)
    #but it looks better if u return table_std

    for i in range(data.shape[1]):
        table_std[:,i] = (data[:,i] - mean_vec[i]*np.ones(data.shape[0])) #take each column and take mean
    return table_std
meancenter = remove_mean(justtwo)
meancenter
#remove_mean(justtwo).shape #test check works!
```

## Question 1 - Unit Vectors, Directions and Projections

#### Part A - Scratch work

Feel free to use this space for scratch work and/or testing for part A

```python
#get vector of len 1 in the right direction:
#aka multiply vector by 1/len of vector (lennorm)
def unit_direction(vector):
    lennorm = np.linalg.norm(vector)
    result = (1/lennorm)*vector
    return result
data_mc = unit_direction(justtwo)
data_mc
```

#### Part B - Scratch work

Feel free to use this space for scratch work and/or testing for part B

```python
#len of unit is dot product
#multiply len by red vector
#len if blue gets projected onto red
def project_datapoint(point, vector): #the blue point is actually a vector already (easier for us to code)
    projectionlen = (np.dot(point, vector))/1 #len of red vector is 1
    if np.linalg.norm(vector) == 1:
        vecdir = unit_direction(vector)
        newvector = vecdir*projectionlen
        return newvector
    else:
        return 'not unit vector'
```

```python
# basically MSE by hand
from scipy.spatial import distance
def mse_n_dim(data_mc, data_rep):
    mse = (distance.cdist(data_mc, data_rep, 'euclidean'))**2
    meanmse = np.mean(mse)
    return meanmse
mse = mse_n_dim(meancenter, data_rep)
mse
```
